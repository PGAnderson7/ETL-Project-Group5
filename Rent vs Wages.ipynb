```python
import pandas as pd
from sqlalchemy import create_engine
```

# Extract CSVs into DataFrames

```python
gross_rent_file = "Resources/gross_rent.csv"
gross_rent_df = pd.read_csv(gross_rent_file)
gross_rent_df.head()
```

```python
wages_file = "Resources/wages.csv"
wages_df = pd.read_csv(wages_file)
wages_df.head()
```

### Transform gross_rent DataFrame

```python
# Create a filtered dataframe from specific columns
gross_rent_cols = ["Zip_Code", "State_Name", "City", "Mean", "Median", "Stdev"]
gross_rent_transformed= gross_rent_df[gross_rent_cols].copy()

# Rename the column headers
gross_rent_transformed = gross_rent_transformed.rename(columns={"Zip_Code": "zip_code",
                                                          "State_Name": "state_name",
                                                          "City": "city",
                                                            "Mean": "mean",
                                                            "Median": "median",
                                                            "Stdev": "stdev"})
```

```python
# Clean the data by dropping duplicates and setting the index
gross_rent_transformed.drop_duplicates("zip_code", inplace=True)
gross_rent_transformed.set_index("zip_code", inplace=True)

gross_rent_transformed.head()
```

```python
gross_rent_transformed=gross_rent_transformed.dropna(subset=['mean'])
```

```python
gross_rent_transformed.head()
```

### Transform wages DataFrame

```python
# Create a filtered dataframe from specific columns
wages_cols = ["Zipcode", "State", "City", "TaxReturnsFiled", "EstimatedPopulation", "TotalWages"]
wages_transformed= wages_df[wages_cols].copy()

# Rename the column headers
wages_transformed = wages_transformed.rename(columns={"Zipcode": "zip_code",
                                                          "State": "state",
                                                          "City": "city",
                                                            "TaxReturnsFiled": "tax_returns_filed",
                                                            "EstimatedPopulation": "est_population",
                                                            "TotalWages": "total_wages"})
```

```python
# Clean the data by dropping duplicates and setting the index
wages_transformed.drop_duplicates("zip_code", inplace=True)
wages_transformed.set_index("zip_code", inplace=True)

wages_transformed.head()
```

```python
wages_transformed=wages_transformed.dropna(subset=['total_wages'])
```

```python
wages_transformed.head()
```

### Create database connection

```python
connection_string = "postgres:postgres@localhost:5432/rent_vs_wages_db"
engine = create_engine(f'postgresql://{connection_string}')
```

```python
# Confirm tables
engine.table_names()
```

# Load DataFrames into database

```python
gross_rent_transformed.to_sql(name='gross_rent', con=engine, if_exists='append', index=True)
```

```python
wages_transformed.to_sql(name='wages', con=engine, if_exists='append', index=True)
```
